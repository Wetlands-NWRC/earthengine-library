```python
import pandas as pd

from eelib.scripttools import MoaTable
```

```python
csv = pd.read_csv("../testing_data/development_Phase01_benchmark_samples-benchmark.csv")
csv.head()
```

```python
csv.columns
```

```python
table = MoaTable(
    dfin=csv,
    label='land_cover'
)
```

```python
table.to_csv("../testing_data/moa-scores.csv")
```

```python
predictors = table.extract_by_rank(rank=1)
predictors
```

```python
table.plot_moa_dis(
    predictors=predictors,
    dir="../testing_data/plots"
)
```
